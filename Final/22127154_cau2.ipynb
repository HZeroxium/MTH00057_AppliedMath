# 3. Ý tưởng thực hiện

Trong bài toán này, ta sẽ sử dụng xích Markov để mô hình hóa quá trình xuống cấp về CSVC của các phòng học thuộc tòa nhà I. Gọi $X_t$ là biến ngẫu nhiên mô tả trạng thái của CSVC của các phòng học vào năm thứ $t$. Ta có thể mô tả xích Markov này bằng ma trận chuyển trạng thái $P$ và vector phân phối xác suất ban đầu $\pi_0$. Tập trạng thái của xích Markov là $S = \{1, 2, 3, 4\}$ tương ứng với các trạng thái chất lượng CSVC của các phòng học.

## a. Xây dựng ma trận chuyển trạng thái $P$

Ma trận chuyển trạng thái $P$ là ma trận với $P_{ij}$ là xác suất chuyển từ một trạng thái $j$ sang một trạng thái $i$ trong một bước thời gian. Ta có thể xây dựng ma trận $P$ từ bảng số liệu cho trước như sau bằng 2 cách suy luận ngược và suy luận tiến:

- Suy luận ngược: Bắt đầu từ trạng thái $4$ (CSVC xấu), ta thấy năm 2020 có 1 phòng có trạng thái $4$, tuy nhiên đến năm $2021$ thì có 2 phòng. Suy ra đã có 1 phòng của năm 2020 chuyển từ trạng thái $3$ sang trạng thái $4$. Từ đó suy ra trong 3 phòng ở trạng thái $3$ của năm 2020, chỉ có 2 phòng giữ nguyên trang thái $3$ vào năm 2021, nhưng số lượng phòng ở trạng thái $3$ vào năm 2021 là 5, suy ra có 3 phòng năm 2020 chuyển từ trạng thái $2$ sang trạng thái $3$. Tương tự, ta suy luận ra các buớc chuyển trạng thái khác.
- SUy luận tiến: Bắt đầu từ trạng thái $1$ (CSVC tốt), ta thấy năm 2020 có 34 phòng có trạng thái $1$, tuy nhiên đến năm $2021$ thì số phòng còn lại là 26. Suy ra có 8 phòng chuyển từ trạng thái $1$ sang trạng thái $2$. Vì số phòng ở trạng thái $2$ của năm 2021 chỉ là 17 nên đã có $8 + 12 - 17 = 3$ phòng chuyển từ trạng thái $2$ sang trạng thái $3$. Tương tự, ta suy luận ra các bước chuyển trạng thái khác.

Để tóm tắt, ta có bảng số liệu suy luận sau:

| Trạng thái | Ban đầu (2020) | Đã chuyển | Giữ nguyên | Lúc sau (2021) |
|------------|---------|-----------|---------|--|
| 1 | 34 | 8 | 26 | 26 |
| 2 | 12 | 3 | 9 | 17 |
| 3 | 3 | 1 | 2 | 5 |
| 4 | 1 | 0 | 1 | 2 |

Từ bảng số liệu trên, ta có ma trận chuyển trạng thái $P$ như sau:

$$ P = \begin{bmatrix} \frac{26}{34} & 0 & 0 & 0 \\ \frac{8}{34} & \frac{9}{12} & 0 & 0 \\ 0 & \frac{3}{12} & \frac{2}{3} & 0 \\ 0 & 0 & \frac{1}{3} & 1 \end{bmatrix} $$


## b. Xác định vector phân phối xác suất ban đầu $\pi_0$

Vector phân phối xác suất ban đầu $\pi_0$ là vector xác suất của trạng thái của xích Markov vào thời điểm ban đầu. Ta có thể xác định vector $\pi_0$ từ bảng số liệu cho trước như sau: (số lượng phòng ở trạng thái $i$ của năm 2021)/(tổng số phòng) với $i = 1, 2, 3, 4$.

$$ \pi_0 = \begin{bmatrix} \frac{26}{50} & \frac{17}{50} & \frac{5}{50} & \frac{2}{50} \end{bmatrix}^T $$

## c. Dự báo sự xuống cấp về CSVC của các phòng học trong vòng 15 năm tới

Để dự báo sự xuống cấp về CSVC của các phòng học trong vòng 15 năm tới, ta sẽ sử dụng ma trận chuyển trạng thái $P$ và vector phân phối xác suất ban đầu $\pi_0$ đã xác định ở trên. Ta sẽ tính vector phân phối xác suất của trạng thái của xích Markov vào các năm tiếp theo bằng cách lần lượt nhân ma trận chuyển trạng thái $P$ với vector phân phối xác suất của trạng thái của xích Markov ở năm trước đó. 

**Ví dụ:** Để xác định vector phân phối xác suất của trạng thái của xích Markov vào năm 2022, ta sẽ tính $$\pi_1 = P \cdot \pi_0$$ Tương tự, để xác định vector phân phối xác suất của trạng thái của xích Markov vào năm 2023, ta sẽ tính $$\pi_2 = P \cdot \pi_1 = P^2 \cdot \pi_0$$ Tương tự, với năm 2035, ta sẽ tính $$\pi_{14} = P^{14} \cdot \pi_0$$.

## d. Xác định thời điểm (năm) phải phá hủy để xây lại/trùng tu tòa nhà I

Để xác định thời điểm (năm) phải phá hủy để xây lại/trùng tu tòa nhà I, ta sẽ xác định năm mà tất cả các phòng học đều ở trạng thái CSVC xấu (trạng thái $4$). Để làm điều này, ta sẽ tính số năm $t$ sao cho phần tử thứ $4$ của vector phân phối xác suất của trạng thái của xích Markov vào năm $t$ là $1$.
$$ \pi_t = P^t \cdot \pi_0 = \begin{bmatrix} 0 & 0 & 0 & 1 \end{bmatrix}^T $$

Ta cũng dễ dàng nhận thấy rằng nếu tất cả các phòng học đều ở trạng thái CSVC xấu, thì sau đó tất cả các phòng học sẽ vẫn ở trạng thái CSVC xấu. Vì vậy, $\pi_{t+1} = P \cdot \pi_t = \begin{bmatrix} 0 & 0 & 0 & 1 \end{bmatrix}^T$ nên $\pi_t$ cũng chính là phân phối dừng của xích Markov.

## e. Xác định tuổi thọ của tòa nhà I

Tuổi thọ của tòa nhà I chính là số năm mà tất cả các phòng học đều ở trạng thái CSVC xấu. Từ kết quả ở câu d, ta sẽ xác định tuổi thọ của tòa nhà I bằng cách lấy số năm $t$ tìm được trừ đi năm 2021.

## 4.2. Xây dựng ma trận chuyển trạng thái $P$ và vector phân phối xác suất ban đầu $\pi_0$

Trong phần này, em sẽ xây dựng ma trận chuyển trạng thái $P$ và vector phân phối xác suất ban đầu $\pi_0$ từ bảng số liệu cho trước. Chi tiết cách xây dựng ma trận chuyển trạng thái và vector phân phối xác suất ban đầu đã được trình bày ở phần 3. Bên dưới là hàm `calculate_transition_matrix_and_initial_distribution` kèm mô tả chi tiết của hàm.

```python
def calculate_transition_matrix_and_initial_distribution(
    data: Matrix,
) -> tuple[Matrix, Matrix, int]:
    """
    Tính toán ma trận chuyển trạng thái và vector phân phối xác suất ban đầu.

    Args:
        data (Matrix): Ma trận dữ liệu đầu vào, mỗi hàng đại diện cho một trạng thái với
                       số lượng phòng học ở hai thời điểm khác nhau.

    Returns:
        tuple[Matrix, Matrix, int]: Trả về ma trận chuyển trạng thái P, vector phân phối xác suất ban đầu pi_0,
                                    và tổng số phòng học.
    """
    num_states = data.rows

    # Khởi tạo ma trận P với các giá trị bằng 0
    P_data = [[0] * num_states for _ in range(num_states)]

    # Tính toán xác suất chuyển trạng thái cho mỗi trạng thái
    for i in range(num_states):
        if i == 0:
            P_data[i][i] = data[i][1] / data[i][0]  # Giữ nguyên trạng thái đầu tiên
            if num_states > 1:
                P_data[i + 1][i] = (data[i][0] - data[i][1]) / data[i][
                    0
                ]  # Chuyển từ trạng thái đầu tiên sang trạng thái tiếp theo
        elif i < num_states - 1:
            rooms_down = sum([data[j][0] - data[j][1] for j in range(i)]) + (
                data[i][0] - data[i][1]
            )
            P_data[i][i] = (data[i][0] - rooms_down) / data[i][
                0
            ]  # Giữ nguyên trạng thái hiện tại
            P_data[i + 1][i] = (
                rooms_down / data[i][0]
            )  # Chuyển xuống trạng thái tiếp theo
        else:
            P_data[i][i] = 1  # Trạng thái cuối cùng, chỉ có thể giữ nguyên

    P = Matrix(P_data)

    # Tính vector phân phối xác suất ban đầu pi_0
    total_rooms = sum([data[i][1] for i in range(num_states)])
    pi_0_data = [[data[i][1] / total_rooms] for i in range(num_states)]
    pi_0 = Matrix(pi_0_data)

    return P, pi_0, total_rooms


# Tính toán ma trận chuyển trạng thái P và vector phân phối xác suất ban đầu pi_0
P, pi_0, total_rooms = calculate_transition_matrix_and_initial_distribution(data)

# In kết quả
print("Ma trận chuyển trạng thái P:")
print(P)
print("\nVector phân phối xác suất ban đầu pi_0:")
print(pi_0)
```

**Chú thích:**

Một trong những cách phổ biến để làm điều này là sử dụng thuật toán làm tròn với trọng số (**Weighted Rounding Algorithm**), cụ thể là phương pháp **Largest Remainder Method** (phương pháp phần dư lớn nhất). Tuy nhiên, dù không gây ra sai số ở tổng số phòng học, phương pháp này có thể tạo ra sai số ở số lượng phòng học ở từng trạng thái, điều này em tạm thời chấp nhận để đơn giản hóa bài toán và tập trung vào mục tiêu chính là dự báo sự xuống cấp về CSVC của các phòng học.

  - Làm tròn xuống cho tất cả các giá trị.

  - Phân bổ phần dư: Xác định các trạng thái có phần dư lớn nhất và phân bổ các phần dư đó cho đến khi tổng số phòng khớp với `total_rooms`.

Bên dưới là các hàm `forecast` kèm mô tả chi tiết của hàm.



```python
def forecast_degradation(P: Matrix, pi_0: Matrix, years: int = 15) -> pd.DataFrame:
    """
    Dự báo sự xuống cấp của CSVC trong vòng 'years' năm tới.

    Args:
        P (Matrix): Ma trận chuyển trạng thái.
        pi_0 (Matrix): Vector phân phối xác suất ban đầu.
        years (int): Số năm để dự báo (mặc định là 15).

    Returns:
        pd.DataFrame: Bảng chứa phân phối xác suất cho từng năm.
    """
    forecast = []
    pi_t = pi_0
    for t in range(years + 1):  # Bao gồm năm hiện tại
        forecast.append([round(x[0], 4) for x in pi_t.data])
        pi_t = P * pi_t

    # Tạo DataFrame để hiển thị kết quả
    years_list = [init_year + i for i in range(years + 1)]
    df_forecast = pd.DataFrame(
        forecast, columns=[f"Trạng thái {i+1}" for i in range(P.rows)], index=years_list
    )

    return df_forecast


def forecast_degradation_integer(
    P: Matrix, pi_0: Matrix, total_rooms: int, years: int = 15
) -> pd.DataFrame:
    """
    Dự báo sự xuống cấp của CSVC trong vòng 'years' năm tới với dữ liệu ở dạng số nguyên.

    Args:
        P (Matrix): Ma trận chuyển trạng thái.
        pi_0 (Matrix): Vector phân phối xác suất ban đầu.
        total_rooms (int): Tổng số phòng học.
        years (int): Số năm để dự báo (mặc định là 15).

    Returns:
        pd.DataFrame: Bảng chứa phân phối xác suất cho từng năm dưới dạng số nguyên.
    """
    forecast = []
    pi_t = pi_0

    for t in range(years + 1):  # Bao gồm năm hiện tại
        # Nhân phân phối xác suất với tổng số phòng học để có số phòng dự đoán
        raw_forecast = [x[0] * total_rooms for x in pi_t.data]

        # Làm tròn xuống và tính phần dư
        rounded_forecast = [int(x) for x in raw_forecast]
        remainders = [x - r for x, r in zip(raw_forecast, rounded_forecast)]

        # Tính tổng sau khi làm tròn
        total_after_rounding = sum(rounded_forecast)

        # Phân bổ phần dư cho các trạng thái có phần dư lớn nhất
        while total_after_rounding < total_rooms:
            max_index = remainders.index(max(remainders))
            rounded_forecast[max_index] += 1
            remainders[max_index] = 0  # Đặt phần dư về 0 để không phân bổ thêm nữa
            total_after_rounding += 1

        forecast.append(rounded_forecast)
        pi_t = P * pi_t

    # Tạo DataFrame để hiển thị kết quả
    years_list = [init_year + i for i in range(years + 1)]
    df_forecast = pd.DataFrame(
        forecast, columns=[f"Trạng thái {i+1}" for i in range(P.rows)], index=years_list
    )

    return df_forecast


def plot_forecast(df_forecast: pd.DataFrame) -> None:
    """
    Vẽ biểu đồ dự báo sự xuống cấp của CSVC.

    Args:
        df_forecast (pd.DataFrame): DataFrame chứa kết quả dự báo.
    """
    df_forecast.plot(kind="line", figsize=(10, 6), marker="o")
    plt.title("Dự báo sự xuống cấp của CSVC các phòng học tòa nhà I")
    plt.xlabel("Năm")
    plt.ylabel("Xác suất trạng thái")
    plt.grid(True)
    plt.show()


# Sử dụng hàm để dự báo và vẽ biểu đồ
df_forecast_float = forecast_degradation(P, pi_0, years=15)
df_forecast_int = forecast_degradation_integer(P, pi_0, total_rooms, years=15)
print("\nBảng dự báo sự xuống cấp của CSVC (dạng xác suất):")
print(df_forecast_float)
plot_forecast(df_forecast_float)
print("\nBảng dự báo sự xuống cấp của CSVC (dạng số nguyên):")
print(df_forecast_int)
plot_forecast(df_forecast_int)
```

## 4.4. Xác định thời điểm (năm) phải phá hủy để xây lại/trùng tu tòa nhà I và tuổi thọ của tòa nhà I

Trong phần này, em sẽ xác định thời điểm (năm) phải phá hủy để xây lại/trùng tu tòa nhà I và tuổi thọ của tòa nhà I từ ma trận chuyển trạng thái $P$ và vector phân phối xác suất ban đầu $\pi_0$. Chi tiết cách xác định đã được trình bày ở phần 3. Bên dưới là các hàm `calculate_destruction_year` và `calculate_lifetime` kèm mô tả chi tiết của hàm.

**Chú thích:**: Trong hàm `find_rebuild_year` có tham số `precision` để xác định độ chính xác của kết quả (mặc định là 4). Thầy cô có thể tùy chỉnh giá trị của `precision` để kiểm tra kết quả với độ chính xác khác nhau.

```python
precision: int = 4


def find_rebuild_year(
    P: Matrix,
    pi_0: Matrix,
    total_rooms: int,
    init_year: int = 2021,
    precision: int = 4,
) -> int:
    """
    Xác định thời điểm cần phá hủy/trùng tu tòa nhà khi tất cả phòng học đều ở trạng thái xấu.

    Args:
        P (Matrix): Ma trận chuyển trạng thái.
        pi_0 (Matrix): Vector phân phối xác suất ban đầu.
        total_rooms (int): Tổng số phòng học.
        start_year (int): Năm bắt đầu theo dõi (mặc định là 2021).
        precision (int): Số chữ số thập phân để so sánh (mặc định là 4).

    Returns:
        int: Năm cần phá hủy/trùng tu.
    """
    pi_t = pi_0
    year = init_year
    while True:
        if (
            round(pi_t[-1][0] * total_rooms, ndigits=precision) == total_rooms
        ):  # Tất cả phòng học đều ở trạng thái 4
            return year
        pi_t = P * pi_t
        year += 1


# Sử dụng hàm để xác định năm cần phá hủy/trùng tu
rebuild_year = find_rebuild_year(
    P, pi_0, total_rooms, init_year=init_year, precision=precision
)
print(f"d. Năm cần phá hủy/trùng tu tòa nhà: {rebuild_year}")


def calculate_lifetime(rebuild_year: int, start_year: int = 2000) -> int:
    """
    Xác định tuổi thọ của tòa nhà I.

    Args:
        rebuild_year (int): Năm phá hủy/trùng tu.
        start_year (int): Năm bắt đầu theo dõi (mặc định là 2021).

    Returns:
        int: Tuổi thọ của tòa nhà I.
    """
    return rebuild_year - start_year


# Sử dụng hàm để xác định tuổi thọ của tòa nhà
lifetime = calculate_lifetime(rebuild_year, start_year=start_year)
print(f"e. Tuổi thọ của tòa nhà I: {lifetime} năm")
```
